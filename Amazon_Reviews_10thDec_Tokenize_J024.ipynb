```
import pandas as pd
import gzip

df=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
 
```

```
import nltk
```

```
from nltk import word_tokenize, pos_tag
from nltk.tokenize import MWETokenizer
from nltk.stem import WordNetLemmatizer
```

```
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
lemmatizer=WordNetLemmatizer()
```

```
def nltkfunc(text):
  nltkdict={'tokenize':[],'pos_tag':[],'lemmatize':[]}
  tokens=nltk.word_tokenize(text)
  nltkdict['tokenize']=tokens
  nltkdict['pos_tag']=nltk.pos_tag(text.split())
  nltkdict['lemmatize']=[lemmatizer.lemmatize(word) for word in tokens]
  return nltkdict
```

```
nltkfunc(df['reviewText'][0])
```

```
df['nltk']=df['reviewText'].apply(nltkfunc)
```

```
def adjective(text):
  x=dict(nltk.pos_tag(text.split()))
  adj=[word for word in x if x[word].startswith('JJ')]
  return adj
```

```
df['reviewTextAdjectives']=df['reviewText'].apply(adjective)
```
