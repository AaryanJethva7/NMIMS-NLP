```
import pandas as pd
import gzip

df=pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz",lines=True)
 
```

```
import re
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def punc(str):
  no_punct = ""
  for char in str:
    if char not in punctuations:
      no_punct = no_punct + char
  return no_punct
```

```
def notBG(word):
  words=re.findall(r'\b(?!b|B)\w+(?<!g)\b',word)
  return words
```

```
df['reviewText'][1] 
```

```
notBG(df['reviewText'][1])
```

```
df['reviewText_not_b_and_g']=df['reviewText'].apply(punc)
df['reviewText_not_b_and_g']=df['reviewText'].apply(notBG)
df['summary_not_b_and_g']=df['summary'].apply(punc)
df['summary_not_b_and_g']=df['summary'].apply(notBG)
```
